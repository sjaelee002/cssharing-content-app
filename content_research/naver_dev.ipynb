```python
from pathlib import Path
from dotenv import load_dotenv
import os
import json
import requests
```

```python
def find_project_root(start_path=None):
    path = Path(start_path or Path.cwd()).resolve()
    for parent in [path] + list(path.parents):
        if (parent / ".env").exists():
            return parent
    raise FileNotFoundError("Could not find .env in current or parent directories.")
```

```python
PROJECT_ROOT = find_project_root()
load_dotenv(PROJECT_ROOT / ".env")
```

```python
NAVER_CLIENT_ID = os.getenv("NAVER_CLIENT_ID")
NAVER_CLIENT_SECRET = os.getenv("NAVER_CLIENT_SECRET")

#print("PROJECT_ROOT:", PROJECT_ROOT)
print("NAVER_CLIENT_ID exists:", bool(NAVER_CLIENT_ID))
print("NAVER_CLIENT_SECRET exists:", bool(NAVER_CLIENT_SECRET))
```
